```python
배열 = [-1, 10.2,"이름", [1,2,3]] # 배열은 순서를 가지고 있다. 첫 순서는 0부터 시작해서 0, 1, 2, 3, 4 순으로 늘어난다.
print(배열)
```

```python
print(배열[0])
print(배열[2])
```

```python
print(배열[3])
```

```python
print(배열[1:3]) # x:y x부터 y전까지(y포함 x) x자리에 아무값도 없으면 "처음부터"를 의미
```

```python
배열[-1] = 10
print(배열)
```
